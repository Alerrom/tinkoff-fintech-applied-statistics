```python
import numpy as np
import pandas as pd
from scipy.stats import uniform, binom, norm, kstest
import scipy.stats as st
%pylab inline
```

## Тест Колмогорова-Смирнова 
H0: x взята из генеральной совокупности с распределением F

H1: x взята из генеральной совокупности с другим распределением

```python
np.random.seed(42)
```

```python
x = st.norm.rvs(size=1000, loc = 0, scale = 1)
```

Сравним нормальное с нормальным

```python
kstest(x, "norm")
```

Сравним нормальное с экспоненциальным

```python
kstest(x, "expon")
```

Сравним нормальное с биномиальным

```python
kstest(x, "binom", args=(1500, 0.16))
```

## QQ-plot

```python
st.probplot(x, dist="norm", plot=pylab);
```

```python
y = st.expon.rvs(size=1500, loc = 0, scale = 1)
st.probplot(y, dist="norm", plot=pylab);
```

```python
z = st.binom.rvs(size=1500, n=1500, p=0.16)
st.probplot(z, dist="norm", plot=pylab);
```
